```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Matplotlib Basics

x = np.linspace(0, 10, 100)
y = np.sin(x)

# Create a simple line graph
plt.plot(x, y)

# Add title and labels
plt.title("Sine Wave")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

```python
# Accessing Subplots 1
plt.subplot(2,2,1)
```

```python
# Subplots
plt.subplot(2,2,1)
plt.plot(x, np.sin(x))
# plt.title("Sine Wave")
plt.subplot(2,2,2)
plt.plot(x, np.cos(x))
# plt.title("Cosine Wave")
plt.subplot(2,2,3)
plt.plot(x, np.tan(x))
# plt.title("Tangent Wave")
plt.subplot(2,2,4)
plt.plot(x, np.exp(-x))
# plt.title("Exponential Decay")
```

```python
# Object-Oriented Approach

# Figsize creates a figure with specified width and height
# DPI sets the resolution of the figure
fig = plt.figure(figsize=(6,6), dpi=100)
fig
# adding axes
# axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height]
# axis1

# adding second axes
axis1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])  # [left, bottom, width, height]
axis1.plot(x, y)

axis2 = fig.add_axes([0.2, 0.5, 0.4, 0.3])
axis2.plot(x, np.cos(x))
```

```python
# Types of Plots
# Scatter Plot
plt.scatter(x, y)
```

```python
# Histogram
data = np.random.randn(1000)
plt.hist(data)
```

```python
# Box Plot
plt.boxplot(data)
```

```python
# Bar Plot
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 12]
plt.bar(categories, values)
```

```python
# Saving Plots
fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x, y)
# save the plot as a PNG file
plt.savefig("plot.png")
```
