```python
# Q. 3 Perform the following operations on advertising_sales.csv:
```

```python
# 1. Load the CSV file and display the first 5 records.
df = pd.read_csv('advertising_sales_extended.csv')
df.head(5)
```

```python
# 2. Consider Advertising_Budget as the input (independent) variable and Sales as the output (dependent) variable.
x = df['Advertising_Budget']
y = df['Sales']
```

```python
# 3. Display summary statistics of both input and output variables.
x.describe() , y.describe()
```

```python
# 4. Draw a scatter plot using Advertising_Budget (input) and Sales (output).
plt.scatter(x, y)
```

```python
# 5. Split the dataset into training and testing sets.
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# 6. Train a Simple Linear Regression model using Advertising_Budget as input and Sales as output.
model = LinearRegression()
model.fit(x_train.values.reshape(-1, 1), y_train)
```

```python
# 7. Display the slope (coefficient) and intercept of the model.
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
# 8. Predict Sales (output) for test data and display actual vs predicted values.
y_pred = model.predict(x_test.values.reshape(-1, 1))
print("Actual vs Predicted Sales:")
print("Actual:", y_test.values)
print("Predicted:", y_pred)
```

```python
# 9. Draw the regression line along with actual data points.
plt.scatter(x, y, color='blue', label='Actual Data')
```

```python
# 10. Predict Sales (output) when Advertising_Budget (input) = 75.
model.predict([[75]])
```

```python
# Q. Perform the following operations on experience_salary.csv:
```

```python
# 1. Load the CSV file and display the last 5 records.
df = pd.read_csv('experience_salary_extended.csv')
df.tail(5)
```

```python
# 2. Consider Experience_Years as the input (independent) variable and Salary as the output (dependent) variable.
x = df['Experience_Years']
y = df['Salary']

print("Experience_Years:", x)
print("Salary:", y)
```

```python
# 3. Check for missing values in both input and output variables.
print("Missing values in Experience_Years:", x.isnull().sum())
print("Missing values in Salary:", y.isnull().sum())
```

```python
# 4. Visualize the relationship using a scatter plot of Experience_Years (input) vs Salary (output).
plt.scatter(x, y, color='blue', label='Actual Data')
plt.xlabel('Experience_Years')
plt.ylabel('Salary')
plt.title('Experience vs Salary')
plt.legend()
plt.show()
```

```python
# 5. Split the dataset into training and testing datasets.
# Correct order: x_train, x_test, y_train, y_test = train_test_split(...)
x_train, x_test, y_train, y_test = train_test_split(x.values, y.values, test_size=0.2, random_state=42)
```

```python
# 6. Build a Simple Linear Regression model using Experience_Years as input and Salary as output
model = LinearRegression()
model.fit(x_train.reshape(-1, 1), y_train)
```

```python
# 7. Display the model coefficient (slope) and intercept.
print("Slope (Coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
# 8. Predict Salary (output) for the test dataset and compare with actual values.
model.predict(x_test.reshape(-1, 1))
```

```python
# 9. Predict Salary (output) for an employee with 8 years of experience (input).
model.predict([[8]])
```
