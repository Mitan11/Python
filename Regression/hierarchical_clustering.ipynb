```python
# ---------------------------------------------------------
# 1. IMPORT REQUIRED LIBRARIES
# ---------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# Library for hierarchical clustering
import scipy.cluster.hierarchy as sch

# Clustering algorithm
from sklearn.cluster import AgglomerativeClustering


# ---------------------------------------------------------
# 2. LOAD DATASET
# ---------------------------------------------------------

data = pd.read_csv("customer_hierarchical_dataset.csv")

print("Dataset Preview:")
print(data.head())


# ---------------------------------------------------------
# 3. SELECT FEATURES
# ---------------------------------------------------------

# Using Annual Income and Spending Score
X = data[['Annual_Income','Spending_Score']]


# ---------------------------------------------------------
# 4. CREATE DENDROGRAM
# ---------------------------------------------------------

plt.figure(figsize=(10,6))

dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))

plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")

plt.show()


# ---------------------------------------------------------
# 5. APPLY HIERARCHICAL CLUSTERING
# ---------------------------------------------------------

model = AgglomerativeClustering(n_clusters=5)

clusters = model.fit_predict(X)

# Add cluster column
data['Cluster'] = clusters

print("\nClustered Data:")
print(data.head())


# ---------------------------------------------------------
# 6. VISUALIZE CLUSTERS
# ---------------------------------------------------------

plt.figure(figsize=(8,6))

plt.scatter(data['Annual_Income'], data['Spending_Score'], c=data['Cluster'])

plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.title("Hierarchical Clustering Result")

plt.show()
```
