```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv('Housing.csv')

x = df[['area']]
y = df['price']
```

```python
plt.scatter(x, y)
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs Price")
plt.show()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

model = LinearRegression()
model.fit(x_train, y_train)
```

```python
print("\nModel Coefficient (Slope):", model.coef_[0])
print("Model Intercept:", model.intercept_)

y_pred = model.predict(x_test)
print("\nActual Price:", y_test)
print("Predicted Price:", y_pred)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("\nMean Squared Error:", mse)
print("R2 Score:", r2)
```

```python
plt.scatter(x, y, label="Actual Data",c='blue')
plt.plot(x, model.predict(x), label="Regression Line", c='red')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Linear Regression Model")
plt.legend()
plt.show()
```

```python
new_hours = np.array([[12]])
predicted_marks = model.predict(new_hours)
print("\nPredicted marks for 12 study hours:", predicted_marks[0])
```
