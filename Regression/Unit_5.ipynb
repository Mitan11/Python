```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: Create dataset
data = {
    'Movie': ['Avengers', 'Batman', 'Titanic', 'Notebook','La La Land'],
    'Genre': ['Action Adventure', 'Action', 'Romance Drama', 'Romance','Romance']
}

df = pd.DataFrame(data)

# Step 2: Convert text (Genre) into numbers
cv = CountVectorizer()
matrix = cv.fit_transform(df['Genre'])

# Step 3: Calculate similarity
similarity = cosine_similarity(matrix)

# Step 4: Recommendation function
def recommend(name):
    # Function to recommend movies similar to the given movie name

    idx = df[df['Movie'] == name].index[0]
    # Find the index of the movie in the dataframe
    # df['Movie'] == name → finds the row where movie name matches
    # .index[0] → gets the index number of that movie

    for i in range(len(df)):
        # Loop through all movies in the dataset

        if i != idx and similarity[idx][i] > 0:
            # Condition 1: i != idx → skip the same movie
            # Condition 2: similarity[idx][i] > 0 → check if movies are similar

            print(df['Movie'][i])
            # Print the name of similar movie

# Step 5: Test
recommend('Notebook')
```

```python
#Collaborative Filtering (User-Based)
'''
| User  | Pizza | Burger | Pasta | Dosa |
| ----- | ----- | ------ | ----- | ---- |
| UserA | 5     | 4      | 1     | 0    |
| UserB | 4     | 0      | 1     | 0    |
| UserC | 0     | 0      | 0     | 5    |
'''
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity

# Step 1: User-Food rating matrix
data = {
    'Pizza': [5, 4, 0],
    'Burger': [4, 0, 0],
    'Pasta': [1, 1, 0],
    'Dosa': [0, 0, 5]
}

df = pd.DataFrame(data, index=['UserA', 'UserB', 'UserC'])

# Step 2: Find similarity between users
similarity = cosine_similarity(df)

sim_df = pd.DataFrame(similarity, index=df.index, columns=df.index)

print("User Similarity Matrix:\n", sim_df)

# Step 3: Recommendation
print("\nRecommended for UserB:")

# Loop through items liked by UserA (similar user)
for item in df.columns:   #Items: Pizza, Burger, Pasta, Dosa
    if df.loc['UserA', item] > 3 and df.loc['UserB', item] == 0:
        # Did UserA like the item? If UserA liked item (rating >3)
        # AND UserB has not rated it
       print(item)
```

```python
# Dimensionality Reduction
# (A) Feature Extraction – PCA(Principal Component Analysis)

'''
Basic Idea
Feature Selection → Choose best features
Feature Extraction → Create new features

Problem:
We have multiple features of students, but we want to reduce features while keeping important information.

=> PCA is used to reduce the number of features by transforming original variables
into new principal components while preserving maximum variance.

Dataset:  StudyHours, Attendance, Assignments
=> Reduce to 2 features using PCA

'''
from sklearn.decomposition import PCA
import pandas as pd

# Step 1: Create dataset (Student performance)
data = {
    'StudyHours': [2, 4, 6, 8],
    'Attendance': [60, 70, 80, 90],
    'Assignments': [1, 2, 3, 4]
}

df = pd.DataFrame(data)

print("Original Data:\n", df)

# Step 2: Apply PCA (reduce 3 features → 2)
pca = PCA(n_components=2)

reduced = pca.fit_transform(df)

# Step 3: Convert into DataFrame
reduced_df = pd.DataFrame(reduced, columns=['PC1', 'PC2'])

print("\nReduced Data (After PCA):\n", reduced_df)


'''
Reduce complexity
Remove redundancy
Improve model performance


Real-life Use Cases:
Image compression
Face recognition
ML model optimization
Big dataset handling
'''
```

```python
#(B) Feature Selection
'''
What is “Best 2 Features”?
It means: Out of A, B, C → only 2 most important features are selected
So:
- Model will use only those 2 features
- Ignore the less important one

Why do we do this?
Because:

- Too many features → slow model
- Some features are useless
- Reduces overfitting

'''
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

# Sample dataset
X = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [2, 3, 4, 5],
    'C': [10, 20, 30, 40]
})

y = [0, 1, 0, 1]

# Select top 2 features
selector = SelectKBest(score_func=f_classif, k=2)
X_new = selector.fit_transform(X, y)

print("Selected Features:\n", X_new)
```
